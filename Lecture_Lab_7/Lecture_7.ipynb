### Implementation

```python
import numpy as np
import matplotlib.pyplot as plt
import numba as nb
from numba import njit
```

```python
kmin = 800
kmax = 1000
penergy_path_gas = 'NSL_SIMULATOR/Exercise_7.2/Gas/OUTPUT/potential_energy.dat'
penergy_path_solid = 'NSL_SIMULATOR/Exercise_7.2/Solid/OUTPUT/potential_energy.dat'
penergy_path_liquid = 'NSL_SIMULATOR/Exercise_7.2/Liquid/OUTPUT/potential_energy.dat'

p_energy_data_gas= np.loadtxt(penergy_path_gas)
p_energy_data_solid= np.loadtxt(penergy_path_solid)
p_energy_data_liquid= np.loadtxt(penergy_path_liquid)

fig, ax1 = plt.subplots(figsize=(16, 10))
colors = ['darkblue', 'darkorchid', 'orchid']
colors_2 = ['darkred', 'red', 'tomato']


ax1.plot(p_energy_data_gas[kmin:kmax, 0], p_energy_data_gas[kmin:kmax, 1], '.-', color=colors[0], label='Gas')
ax1.plot(p_energy_data_solid[kmin:kmax, 0], p_energy_data_solid[kmin:kmax, 1], '.-', color=colors[1], label='Solid')
ax1.plot(p_energy_data_liquid[kmin:kmax, 0], p_energy_data_liquid[kmin:kmax, 1],'.-', color=colors[2], label='Liquid')
ax1.axhline(y=p_energy_data_gas[-1,1], color='r', linestyle='--', label='Average', linewidth=1)
ax1.axhline(y=p_energy_data_solid[-1,1], color='r', linestyle='--', linewidth=1)
ax1.axhline(y=p_energy_data_liquid[-1,1], color='r', linestyle='--', linewidth=1)

ax1.set_xlabel("Block", fontsize=14)
ax1.set_ylabel("U / N", fontsize=14)
ax1.set_title('Potential Energy Dynamics of Monte Carlo Simulation', fontsize=30, fontweight='bold')
ax1.grid(True)
ax1.legend(loc='center right', fontsize=12)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
# parallelized compiled function 
@nb.njit("float64[:](float64[:], int64)", parallel=True)
def compute_autocorrelation(data, max_lag):
    N = len(data)
    mean_data = np.mean(data)
    variance_data = np.var(data)
    autocorr = np.zeros(max_lag)
    
    # parallelized loop
    for tau in nb.prange(1, max_lag + 1):
        s1 = 0.0
        s2 = 0.0

        for tp in range(N - tau):
            s1 += (data[tp] - mean_data) * (data[tp + tau] - mean_data)
            s2 += (data[tp] - mean_data) ** 2

        autocorr[tau - 1] = s1 / s2
    return autocorr

# maximum number of steps for autocorrelation
max_lag = 500  

autocorr_gas = compute_autocorrelation(p_energy_data_gas[:,1], max_lag)
autocorr_liquid = compute_autocorrelation(p_energy_data_solid[:,1], max_lag)
autocorr_solid = compute_autocorrelation(p_energy_data_liquid[:,1], max_lag)

plt.figure(figsize=(16, 10))
plt.plot(np.arange(1, max_lag + 1), autocorr_gas, label='Gas', color='darkblue')
plt.plot(np.arange(1, max_lag + 1), autocorr_liquid, label='Liquid', color='darkorchid')
plt.plot(np.arange(1, max_lag + 1), autocorr_solid, label='Solid', color='orchid')
plt.axvline(x=300, color='r', linestyle='--', linewidth=1)
plt.xlabel("Metropolis Step")
plt.ylabel("Autocorrelation")
plt.legend()
plt.grid(True)
plt.title("Autocorrelation Function", fontsize=25, fontweight="bold")
plt.show()
```

The autocorrelation function becomes approximately constant and approaches zero after about 300 Metropolis steps. Consequently, an appropriate range for the data blocking block size would be from 5 times to 10 times the autocorrelation length, which is 1500 to 3000 steps. A suitable choice within this range is a block size of 2000 steps

```python
@njit("float64[:](float64[:], int64)")
def compute_block_means(data, block_size):
    num_blocks = len(data) // block_size
    block_means = np.empty(num_blocks)
    for i in range(num_blocks):
        block_means[i] = np.mean(data[i * block_size:(i + 1) * block_size])
    return block_means

@njit("Tuple((int64[:], float64[:]))(float64[:], int64, int64)")
def compute_statistical_uncertainties(data, min_block_size, max_block_size):
    num_block_sizes = max_block_size - min_block_size + 1
    block_sizes = np.arange(min_block_size, max_block_size + 1)
    uncertainties = np.empty(num_block_sizes)

    for i in range(num_block_sizes):
        block_size = block_sizes[i]
        block_means = compute_block_means(data, block_size)
        uncertainties[i] = np.sqrt(np.var(block_means) / len(block_means))

    return block_sizes, uncertainties


min_block_size = 10
max_block_size = 5000

block_sizes_gas, uncertainties_gas = compute_statistical_uncertainties(p_energy_data_gas[:,1], min_block_size, max_block_size)
block_sizes_solid, uncertainties_solid = compute_statistical_uncertainties(p_energy_data_solid[:,1], min_block_size, max_block_size)
block_sizes_liquid, uncertainties_liquid = compute_statistical_uncertainties(p_energy_data_liquid[:,1], min_block_size, max_block_size)

plt.figure(figsize=(16, 10))
plt.plot(block_sizes_gas, uncertainties_gas, label='Gas', color='darkblue')
plt.plot(block_sizes_solid, uncertainties_solid, label='Solid', color='darkorchid')
plt.plot(block_sizes_liquid, uncertainties_liquid, label='Liquid', color='orchid')
plt.axvline(x=300, color='r', linestyle='--', linewidth=1, label="300 steps")
plt.axvline(x=2000, color='black', linestyle='--', linewidth=1, label="2000 steps")

plt.xlabel("Block Size")
plt.ylabel("Error")
plt.legend()
plt.grid(True)
plt.title("Statistical Uncertainty", fontsize=25, fontweight="bold")
plt.show()
```

2. liquid phase: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
3. gas phase: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)

show in pictures the obtained average values and uncertainties for the potential energy per particle, $U/N$, the pressure $P$ and the radial distribution function $g(r)$ in LJ units ... <span style="color:red"> and compare your MC results for the radial distribution function, $g(r)$, with those obtained with Molecular Dynamics NVE simulations in similar thermodynamic conditions.</span>

```python
def load_and_plot_data(file_paths, ylabel, title, suptitle, y_lines=None):
    data_labels = ['Gas', 'Solid', 'Liquid']
    colors = ['darkblue', 'darkorchid', 'orchid']

    fig, axs = plt.subplots(1, 3, figsize=(24, 8))  # Create 3 subplots in one row

    for i, file_path in enumerate(file_paths):
        data = np.loadtxt(file_path)
        ax = axs[i]  # Select the subplot
        ax.errorbar(data[:, 0], data[:, 2], yerr=data[:, 3], fmt='o-', color=colors[i], label=data_labels[i], linewidth=2, markersize=5)
        
        if y_lines:
            for y in y_lines:
                ax.axhline(y=y, color='black', linestyle='--', linewidth=2)
        
        ax.set_xlabel("Blocks", fontsize=14)
        ax.set_ylabel(ylabel, fontsize=14)
        ax.set_title(f"{title} ({data_labels[i]})", fontsize=18)
        ax.grid(True)
        ax.legend(loc='best', fontsize=12)

    fig.suptitle(suptitle, fontsize=30, fontweight='bold')
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()
```

```python
penergy_paths = ['NSL_SIMULATOR/Exercise_7.4/MC/Gas/OUTPUT/potential_energy.dat', 
                 'NSL_SIMULATOR/Exercise_7.4/MC/Liquid/OUTPUT/potential_energy.dat', 
                 'NSL_SIMULATOR/Exercise_7.4/MC/Solid/OUTPUT/potential_energy.dat']

load_and_plot_data(penergy_paths, "U/N", "Potential Energy per Particle", "NVT Monte Carlo Simulation")
```
