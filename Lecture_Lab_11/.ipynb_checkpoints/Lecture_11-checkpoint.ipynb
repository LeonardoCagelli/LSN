## Exercise 11.1

In order to make practice with NN, explore how does the previous linear regression depend on the number of epochs, $N_{\mathrm{epochs}}$, the number of data points $N_{\mathrm{train}}$ and on the noise $\sigma$. Try to improve the previous result operating on these parameters.

### Implementation

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import os
```

```python
m = 2
q = 1

def linear_function(x):
    return m*x + q


train_sizes = [100, 200, 300, 400]
test_sizes = [10, 20, 30, 40]
epochs = [10, 20, 30, 40]
colors = ['red', 'yellow', 'brown', 'green']

plt.figure(figsize=(10, 20))

for i in range(4):

    x_train = np.random.uniform(-1, 1, train_sizes[i])
    x_test = np.random.uniform(-1, 1, test_sizes[i])

    y_train = linear_function(x_train) + np.random.normal(0, 0.5, train_sizes[i])
    y_test = linear_function(x_test) + np.random.normal(0, 0.5, test_sizes[i])

    plt.subplot(4, 2, i*2 + 2)
    x = np.linspace(-2, 2, 100)
    y_true = linear_function(x)
    plt.plot(x_train, y_train, '.', color="blue", label='Training data')
    plt.plot(x, y_true, '-', color="black", label="True function")

    for j in range(4):

        # Define the model
        model = tf.keras.Sequential()
        model.add(tf.keras.layers.Dense(1, input_shape=(1,)))
        model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
        history = model.fit(x_train, y_train, batch_size=30, epochs=epochs[j], shuffle=True, validation_data=(x_test, y_test))

        plt.subplot(4, 2, 2*i + 1)
        plt.plot(history.history['val_loss'], label=("Loss with " + str(epochs[j]) + " epochs"), color=colors[j])

        plt.subplot(4, 2, 2*i + 2)
        y_pred = model.predict(x)
        plt.plot(x, y_pred, '--', color=colors[j], label=("Model with " + str(epochs[j]) + " epochs"))

    plt.subplot(4, 2, 2*i+1)
    plt.grid()
    plt.title('Model loss with ' + str(train_sizes[i]) + " points", fontsize=10)
    plt.ylabel('Loss', fontsize=(5))
    plt.xlabel('Epoch', fontsize=(5))
    plt.legend(fontsize=(5))

    plt.subplot(4, 2, 2*i+2)
    plt.title('Resume with ' + str(train_sizes[i]) + " points", fontsize=10)
    plt.grid()
    plt.xlabel('X', fontsize=(5))
    plt.ylabel('Y', fontsize=(5))
    plt.legend(fontsize=(5))

plt.show()
```

## Exercise 11.2

Try to extend the model to obtain a reasonable fit of the following polynomial of order 3:

$$
f(x)=4-3x-2x^2+3x^3
$$
for $x \in [-1,1]$.

Find good (& reasonable) choices for:

- the number of layers
- the number of neurons in each layer
- the activation function
- the optimizer
- the loss function
  
Check your NN model by seeing how well your fits predict newly generated test data (including on data outside the range you fit. How well do your NN do on points in the range of $x$ where you trained the model? How about points outside the original training data set? 
Summarize what you have learned about the relationship between model complexity (number of parameters), goodness of fit on training data, and the ability to predict well.

## Implementation

```python
def f(x):
    return 4 - 3 * x - 2 * x**2 + 3 * x**3

# Create the neural network
def define_model(n_hidden_layers, n_neurons, f_activation, f_optimizer, f_loss):
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Dense(n_neurons, input_shape=(1, ), activation=f_activation))

    for i in range(n_hidden_layers - 1):
        model.add(tf.keras.layers.Dense(n_neurons, activation=f_activation))
    
    model.add(tf.keras.layers.Dense(1, activation='linear'))

    model.compile(optimizer=f_optimizer, loss=f_loss, metrics=f_loss)

    return model

# Generate training and validation dataset
def generate_data(x_min, x_max, dim_train, dim_val, function, sigma):
    x_train = np.random.uniform(x_min, x_max, dim_train)
    x_valid = np.random.uniform(x_min, x_max, dim_val)

    y_train = function(x_train) + np.random.normal(0, sigma, len(x_train))
    y_valid = function(x_valid) + np.random.normal(0, sigma, len(x_valid))

    return x_train, y_train, x_valid, y_valid
```

```python
x_train, y_train, x_valid, y_valid = generate_data(-1, 1, 1000, 100, f, 0.5)
x = np.linspace(-1.2, 1.2, 100)
y_true = f(x)
```

### Test the loss function performance

```python
loss_functions = ['MSE', 'MAE']
colors = ['darkviolet', 'blue', 'plum', 'royalblue']

plt.figure(figsize=(12, 6))
plt.suptitle('Loss function trials:\nMean Squarred Error vs Mean Absolute Error', fontsize=12)
plt.subplot(1, 2, 2)
plt.plot(x_train, y_train, '.', color='tomato', label='Training data')
plt.plot(x, y_true, '--', label='True Function', color='black', linewidth='3.0')

i = 0

for loss_f in loss_functions:
    model = define_model(2, 10, 'relu', 'sgd', loss_f)
    history = model.fit(x_train, y_train, batch_size=30, epochs=40, shuffle=True, validation_data=(x_valid, y_valid))

    validation_loss = history.history['val_loss'][-1]
    y_model = model.predict(x)

    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label=f'Train Loss {loss_f}', color=colors[i])
    plt.plot(history.history['val_loss'], label=f'Validation Loss {loss_f}', color=colors[i+2])
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(x, y_model, '-', label=f'Model Function \nLoss function: {loss_f} \nValidation loss: {validation_loss:.3f}', 
                    color=colors[i], linewidth='3.0')
    plt.legend()
    plt.title('Model Prediction')

    i += 1

plt.subplot(1, 2, 1)
plt.grid()
plt.subplot(1, 2, 2)
plt.grid()
plt.show()
```

The Mean Squared Error (MSE) loss function appears to be significantly more effective for our task. Therefore, I will continue using the MSE loss function
### Test the activation function performance

```python
activation_functions = ['relu', 'sigmoid']
colors = ['darkviolet', 'blue', 'plum', 'royalblue']

plt.figure(figsize=(12, 6))
plt.suptitle('Activation function trials:\nReLU vs Sigmoid', fontsize=12)
plt.subplot(1, 2, 2)
plt.plot(x_train, y_train, '.', color='tomato', label='Training data')
plt.plot(x, y_true, '--', label='True Function', color='black', linewidth='3.0')

i = 0

for activation_f in activation_functions:
    model = define_model(2, 10, activation_f, 'sgd', 'mse')
    history = model.fit(x_train, y_train, batch_size=30, epochs=40, shuffle=True, validation_data=(x_valid, y_valid))

    validation_loss = history.history['val_loss'][-1]
    y_model = model.predict(x)

    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label=f'Train Loss {activation_f}', color=colors[i])
    plt.plot(history.history['val_loss'], label=f'Validation Loss {activation_f}', color=colors[i+2])
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(x, y_model, '-', label=f'Model Function \nActivation function: {activation_f} \nValidation loss: {validation_loss:.3f}', 
                    color=colors[i], linewidth='3.0')
    plt.legend()
    plt.title('Model Prediction')

    i += 1

plt.subplot(1, 2, 1)
plt.grid()
plt.subplot(1, 2, 2)
plt.grid()
plt.show()
```

The ReLU activation function has proven to be significantly more effective for our task. Therefore, I will continue using the ReLU activation function
### Test the optmimizer performance

```python
optimizers = ['SGD', 'ADAM']
colors = ['darkviolet', 'blue', 'plum', 'royalblue']

plt.figure(figsize=(12, 6))
plt.suptitle('Optimizers trials:\nStochastic Gradient Descent vs ADAM', fontsize=12)
plt.subplot(1, 2, 2)
plt.plot(x_train, y_train, '.', color='tomato', label='Training data')
plt.plot(x, y_true, '--', label='True Function', color='black', linewidth='3.0')

i = 0

for opt in optimizers:
    model = define_model(2, 10, 'relu', opt, 'mse')
    history = model.fit(x_train, y_train, batch_size=30, epochs=40, shuffle=True, validation_data=(x_valid, y_valid))

    validation_loss = history.history['val_loss'][-1]
    y_model = model.predict(x)

    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label=f'Train Loss {opt}', color=colors[i])
    plt.plot(history.history['val_loss'], label=f'Validation Loss {opt}', color=colors[i+2])
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(x, y_model, '-', label=f'Model Function \nOptimizer: {opt} \nValidation loss: {validation_loss:.3f}', 
                    color=colors[i], linewidth='3.0')
    plt.legend()
    plt.title('Model Prediction')

    i += 1

plt.subplot(1, 2, 1)
plt.grid()
plt.subplot(1, 2, 2)
plt.grid()
plt.show()
```

The performance of the ADAM and SGD optimizers appears to be similar. However the prediciton shows a more accurate plot of the function trained with SGD. Therefore, I will continue using the SGD optimizer.


### Test the Neural Network Architecture

```python
import pandas as pd 

# Define the list of colors
colors = ['plum', 'violet', 'magenta', 'mediumorchid', 'rebeccapurple']

# File paths for saved data
training_loss_file = 'models/training_loss_units.txt'
validation_loss_file = 'models/validation_loss_units.txt'
units_file = 'models/units.txt'
```

```python
# Check if the data files exist
if os.path.exists(training_loss_file) and os.path.exists(validation_loss_file) and os.path.exists(units_file):
    # Load the data from files
    training_loss = np.loadtxt(training_loss_file)
    validation_loss = np.loadtxt(validation_loss_file)
    v_units = np.loadtxt(units_file)
else:
    # Initialize arrays to store the data
    v_units = np.empty(0)
    training_loss = np.empty(0)
    validation_loss = np.empty(0)

    plt.figure(figsize=(12, 6))
    plt.suptitle('Fitting Quality for increasing units in one hidden layer', fontsize=12)
    plt.subplot(1, 2, 2)
    plt.plot(x_train, y_train, '.', color='tomato', label='Training data')
    plt.plot(x, y_true, '--', label='True Function', color='black', linewidth=3.0)

    # Index for the color
    i = 0 
    for units in range(10, 251, 5):
        model = define_model(1, units, 'relu', 'sgd', 'mse')
        history = model.fit(x_train, y_train, batch_size=30, epochs=40, shuffle=True, validation_data=(x_valid, y_valid))

        validation_loss = np.append(validation_loss, history.history['val_loss'][-1])
        training_loss = np.append(training_loss, history.history['loss'][-1])
        v_units = np.append(v_units, units)

        y_model = model.predict(x)

        if units % 50 == 0:
            plt.subplot(1, 2, 2)
            plt.plot(x, y_model, '-', label=f'Model {units} units: val loss = {validation_loss[-1]:.3f}', linewidth=3.0, color=colors[i])
            model.save(f"models/model_{units}_units.h5")
            i += 1

    # Save the data to files
    np.savetxt(training_loss_file, training_loss, fmt='%.3f')
    np.savetxt(validation_loss_file, validation_loss, fmt='%.3f')
    np.savetxt(units_file, v_units, fmt='%.3f')

plt.figure(figsize=(12, 6))
plt.suptitle('Fitting Quality for increasing units in one hidden layer', fontsize=18)
plt.subplot(1, 2, 2)
plt.plot(x_train, y_train, '.', color='tomato', label='Training data')
plt.plot(x, y_true, '--', label='True Function', color='black', linewidth=3.0)


# Re-plot the saved models if they exist
if os.path.exists(training_loss_file) and os.path.exists(validation_loss_file) and os.path.exists(units_file):
    i = 0
    for units in range(50, 251, 50):
        model = define_model(1, units, 'relu', 'sgd', 'mse')
        model.load_weights(f"models/model_{units}_units.h5")
        y_model = model.predict(x)
        plt.plot(x, y_model, '-', label=f'Model {units} units: val loss = {validation_loss[i]:.3f}', linewidth=3.0, color=colors[i % len(colors)])
        i += 1

plt.subplot(1, 2, 2)
plt.legend()
plt.title('Model Prediction')

plt.subplot(1, 2, 1)
plt.plot(v_units, training_loss, '.-', label='Training Loss', color='darkviolet')
plt.plot(v_units, validation_loss, '.-', label='Validation Loss', color='blue')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Number of Units')
plt.legend()

plt.show()
```

The validation loss remains relatively constant with more than 100 units in the single hidden layer. The minimum of validation loss is reached with 75 units. Increasing the number of units beyond this point is likely to cause overfitting, reducing the model's ability to generalize

```python
# Define colors and units
colors = ['plum', 'violet', 'magenta', 'darkviolet', 'indigo']
units = [20, 25, 30]

plt.figure(figsize=(10, 15))
plt.suptitle('Fitting Quality for Increasing Layers and Units', fontsize=18)
plt.subplots_adjust(top=0.88)  # Adjust top spacing to avoid overlapping with suptitle

for i in range(3):
    plt.subplot(3, 2, 2 + 2*i)
    plt.plot(x_train, y_train, '.', color='tomato', label='Training data')
    plt.plot(x, y_true, '--', label='True Function', color='black', linewidth=3.0)

# Index for the color
j = 1

for unit in units:
    v_layers = np.empty(0)
    training_loss = np.empty(0)
    validation_loss = np.empty(0)

    for layers in range(1, 6):
        model = define_model(layers, unit, 'relu', 'sgd', 'mse')
        history = model.fit(x_train, y_train, batch_size=30, epochs=40, shuffle=True, validation_data=(x_valid, y_valid), verbose=0)

        validation_loss = np.append(validation_loss, history.history['val_loss'][-1])
        training_loss = np.append(training_loss, history.history['loss'][-1])
        v_layers = np.append(v_layers, layers)

        y_model = model.predict(x)

        plt.subplot(3, 2, 2*j)
        plt.plot(x, y_model, '-', label=f'Model {layers} | {unit}: val loss = {validation_loss[-1]:.3f}', linewidth=(3.0 - layers/2.5), color=colors[layers-1])
        model.save(f"models/model_{layers}_layers_{unit}_units.h5")

    np.savetxt(f'models/training_loss_layers_{unit}.txt', training_loss, fmt='%.3f')
    np.savetxt(f'models/validation_loss_layers_{unit}.txt', validation_loss, fmt='%.3f')
    np.savetxt(f'models/layers.txt', v_layers, fmt='%.3f')

    plt.subplot(3, 2, 2*j)
    plt.legend(fontsize=7.0)
    plt.title(f'Model Prediction {unit} units', fontsize=10)
    plt.grid()

    plt.subplot(3, 2, 2*j-1)
    plt.plot(v_layers, training_loss, '.-', label=f'Training Loss {unit} units', color='darkviolet')
    plt.plot(v_layers, validation_loss, '.-', label=f'Validation Loss {unit} units', color='blue')
    plt.title(f'Model Loss {unit} units', fontsize=10)
    plt.ylabel('Loss', fontsize=6.0)
    plt.xlabel('Number of layers', fontsize=6.0)
    plt.legend(fontsize=7.0)
    plt.grid()

    j += 1

# Adjust layout to avoid overlapping titles and labels
plt.tight_layout(rect=[0, 0, 1, 0.95])

plt.show()
```

The best validation loss is achieved with a network of 3 hidden layers, each containing 25 units, totaling 75 neurons. Curiously, this same result was previously obtained with a single hidden layer of 75 units. However, the validation loss is smaller in the latter case (0.253 vs. 0.281).

## Exercise 11.3

Try to extend the model to fit a simple trigonometric 2D function such as $f(x,y) = \sin(x^2+y^2)$ in the range $x \in [-3/2,3/2]$ and $y \in [-3/2,3/2]$.

## Implementation

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
xy_min = [-3/2., -3/2.]
xy_max = [3/2., 3/2.]

# Function to fit
def f(x):
    return np.sin(x[:, 0]**2 + x[:, 1]**2)

# Generate training and validation data
def generate_data(train_size, test_size):
    x_train = np.random.uniform(low=xy_min, high=xy_max, size=(train_size, 2))
    y_train = f(x_train) + np.random.normal(0, 0.1, train_size)
    
    x_valid = np.random.uniform(low=xy_min, high=xy_max, size=(test_size, 2))
    y_valid = f(x_valid) + np.random.normal(0, 0.1, test_size)

    return x_train, y_train, x_valid, y_valid    

# Create the neural network
def define_model(n_hidden_layers, n_neurons, f_activation, f_optimizer, f_loss):
    model = tf.keras.Sequential()

    for i in range(n_hidden_layers):
        if i == 0:
            model.add(tf.keras.layers.Dense(n_neurons, input_shape=(2, ), activation=f_activation))
        else:
            model.add(tf.keras.layers.Dense(n_neurons, activation=f_activation))
    
    model.add(tf.keras.layers.Dense(1, activation='linear'))

    model.compile(optimizer=f_optimizer, loss=f_loss, metrics=[f_loss])

    return model
```

```python
x_train, y_train, x_valid, y_valid = generate_data(2000, 200)

fig = plt.figure(figsize=(15, 5))
ax = fig.add_subplot(131, projection='3d')
ax.scatter(x_train[:,0], x_train[:,1], y_train)
ax.set_title('Training Data')

model = define_model(4, 20, 'relu', 'sgd', 'mse')
model.fit(x_train, y_train, batch_size=30, epochs=60, validation_data=(x_valid, y_valid))

X = np.linspace(xy_min[0], xy_max[0], 100)
Y = np.linspace(xy_min[1], xy_max[1], 100)
X, Y = np.meshgrid(X, Y)
XY_grid = np.c_[X.ravel(), Y.ravel()]

Z_pred = model.predict(XY_grid).reshape(X.shape)
Z_true = f(XY_grid).reshape(X.shape)

ax = fig.add_subplot(1,3,2, projection='3d')
ax.plot_surface(X, Y, Z_true, cmap='viridis', alpha=0.8)
ax.set_title('True Function')

ax = fig.add_subplot(1,3,3, projection='3d')
ax.plot_surface(X, Y, Z_pred, cmap='viridis', alpha=0.8)
ax.set_title('Predicted Function')

plt.show()
```
