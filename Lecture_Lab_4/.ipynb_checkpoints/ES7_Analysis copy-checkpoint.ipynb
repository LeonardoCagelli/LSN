# Lecture 7 Data analysis

### Kinetic Energy Analysis

```python
import numpy as np
import matplotlib.pyplot as plt

# Paths to the data files
kenergy_path_gas = 'NSL_SIMULATOR/ES7/Gas/OUTPUT/kinetic_energy.dat'
kenergy_path_solid = 'NSL_SIMULATOR/ES7/Solid/OUTPUT/kinetic_energy.dat'
kenergy_path_liquid = 'NSL_SIMULATOR/ES7/Liquid/OUTPUT/kinetic_energy.dat'

# Load data
k_energy_data_gas= np.loadtxt(kenergy_path_gas)
k_energy_data_solid= np.loadtxt(kenergy_path_solid)
k_energy_data_liquid= np.loadtxt(kenergy_path_liquid)

plt.style.use('seaborn-darkgrid')
fig, ax1 = plt.subplots(figsize=(16, 10))
colors = ['darkblue', 'darkorchid', 'orchid']

ax1.errorbar(k_energy_data_gas[:, 0], k_energy_data_gas[:, 2], yerr=k_energy_data_gas[:, 3], fmt='o-', color=colors[0], label='Gas', linewidth=2, markersize=5)
ax1.errorbar(k_energy_data_solid[:, 0], k_energy_data_solid[:, 2], yerr=k_energy_data_solid[:, 3], fmt='o-', color=colors[1], label='Solid', linewidth=2, markersize=5)
ax1.errorbar(k_energy_data_liquid[:, 0], k_energy_data_liquid[:, 2], yerr=k_energy_data_liquid[:, 3], fmt='o-', color=colors[2], label='Liquid', linewidth=2, markersize=5)
ax1.set_xlabel("Blocks", fontsize=14)
ax1.set_ylabel("Kinetic Energy per particle", fontsize=14)
ax1.set_title('Kinetic Energy Dynamics', fontsize=18)
ax1.grid(True)
ax1.legend(loc='center right', fontsize=12)
fig.suptitle("Kinetic Energy Simulation Analysis", fontsize=30, fontweight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.95])
```

### Temperature Analysis

```python
import numpy as np
import matplotlib.pyplot as plt

# Paths to the data files
temperature_path_gas = 'NSL_SIMULATOR/ES7/Gas/OUTPUT/temperature.dat'
temperature_path_solid = 'NSL_SIMULATOR/ES7/Solid/OUTPUT/temperature.dat'
temperature_path_liquid = 'NSL_SIMULATOR/ES7/Liquid/OUTPUT/temperature.dat'

# Load data
temperature_data_gas = np.loadtxt(temperature_path_gas)
temperature_data_solid = np.loadtxt(temperature_path_solid)
temperature_data_liquid = np.loadtxt(temperature_path_liquid)

plt.style.use('seaborn-darkgrid')
fig, ax1 = plt.subplots(figsize=(16, 10))
colors = ['darkblue', 'darkorchid', 'orchid']

ax1.errorbar(temperature_data_gas[:, 0], temperature_data_gas[:, 2], yerr=temperature_data_gas[:, 3], fmt='o-', color=colors[0], label='Gas', linewidth=2, markersize=5)
ax1.errorbar(temperature_data_solid[:, 0], temperature_data_solid[:, 2], yerr=temperature_data_solid[:, 3], fmt='o-', color=colors[1], label='Solid', linewidth=2, markersize=5)
ax1.errorbar(temperature_data_liquid[:, 0], temperature_data_liquid[:, 2], yerr=temperature_data_liquid[:, 3], fmt='o-', color=colors[2], label='Liquid', linewidth=2, markersize=5)
plt.axhline(y=0.8, color='black', linestyle='--', linewidth=2)
plt.axhline(y=1.1, color='black', linestyle='--', linewidth=2)
plt.axhline(y=1.2, color='black', linestyle='--', linewidth=2)
ax1.set_xlabel("Blocks", fontsize=14)
ax1.set_ylabel("Temperature", fontsize=14)
ax1.set_title('Temperature Dynamics', fontsize=18)
ax1.grid(True)
ax1.legend(loc='center right', fontsize=12)
fig.suptitle("Temperature Simulation Analysis", fontsize=30, fontweight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.95])
```
