```python
import numpy as np
import matplotlib.pyplot as plt

v_temp = ['0.500000', '0.750000', '1.000000', '1.250000', '1.500000', '1.750000', '2.000000']
sim_type = ['MRT2', 'GIBBS']
points = 100
T = np.linspace(0.2, 3.0, num=points)
beta = 1 / T
J = 1.0
Ns = 50
h = 0.02

th = np.tanh(J / T)
thN = th**Ns
ch = 1 / th
b = 1 / T
l1 = np.exp(b * J) * np.cosh(b * h) + np.sqrt(np.exp(2 * b * J) * np.cosh(b * h)**2 - 2 * np.sinh(2 * b * J))
l2 = np.exp(b * J) * np.cosh(b * h) - np.sqrt(np.exp(2 * b * J) * np.cosh(b * h)**2 - 2 * np.sinh(2 * b * J))
Z = l1**Ns + l2**Ns

M = (np.exp(b * J) * np.sinh(b * h) * 
    ((l1**(Ns - 1)) * (1 + np.exp(b * J) * np.cosh(b * h) / np.sqrt(np.exp(2 * b * J) * np.cosh(b * h)**2 - 2 * np.sinh(2 * b * J))) + 
     (l2**(Ns - 1)) * (1 - np.exp(b * J) * np.cosh(b * h) / np.sqrt(np.exp(2 * b * J) * np.cosh(b * h)**2 - 2 * np.sinh(2 * b * J)))) / Z)
X = beta * np.exp(2 * beta * J) * (1 - thN) / (1 + thN)
heat = ((beta * J)**2) * (((1 + thN + (Ns - 1) * (th**2) + (Ns - 1) * (ch**2) * thN) / (1 + thN)) - Ns * ((th + ch * thN) / (1 + thN))**2)
e = -J * (th + ch * thN) / (1 + thN)

colors = ['darkblue', 'orchid']
markers = ['v', '^']

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 16))

ax1.plot(T, M, '--', color='black', label='Theoretical Value')
ax2.plot(T, X, '--', color='black', label="Theoretical Value")
ax3.plot(T, e, "--", color='black', label="Theoretical Value")
ax4.plot(T, heat, "--", color='black', label="Theoretical Value")

def load_data(sim, temp, metric):
    path = f'NSL_SIMULATOR/ES6/Ising/OUTPUT/{metric}_{sim}_{temp}.dat'
    return np.loadtxt(path)

v_m_mrt2, v_chi_mrt2, v_te_mrt2, v_cv_mrt2 = [], [], [], []
v_m_gibbs, v_chi_gibbs, v_te_gibbs, v_cv_gibbs = [], [], [], []
v_m_mrt2_err, v_chi_mrt2_err, v_te_mrt2_err, v_cv_mrt2_err = [], [], [], []
v_m_gibbs_err, v_chi_gibbs_err, v_te_gibbs_err, v_cv_gibbs_err = [], [], [], []

for c, sim in enumerate(sim_type):
    v_m, v_chi, v_te, v_cv, v_t = [], [], [], [], []
    v_m_err, v_chi_err, v_te_err, v_cv_err = [], [], [], []

    for temp in v_temp:
        magnetization_data = load_data(sim, temp, 'magnetization')
        susceptibility_data = load_data(sim, temp, 'susceptibility')
        total_energy_data = load_data(sim, temp, 'total_energy')
        specific_heat_data = load_data(sim, temp, 'specific_heat')

        v_m.append(magnetization_data[-1, 2])
        v_chi.append(susceptibility_data[-1, 2])
        v_te.append(total_energy_data[-1, 2])
        v_cv.append(specific_heat_data[-1, 2])
        v_t.append(float(temp))
        
        v_m_err.append(magnetization_data[-1, 3])
        v_chi_err.append(magnetization_data[-1, 3])
        v_te_err.append(magnetization_data[-1, 3])
        v_cv_err.append(magnetization_data[-1, 3])

    ax1.plot(v_t, v_m, markers[c], color=colors[c], label=sim)
    ax2.plot(v_t, v_chi, markers[c], color=colors[c], label=sim)
    ax3.plot(v_t, v_te, markers[c], color=colors[c], label=sim)
    ax4.plot(v_t, v_cv, markers[c], color=colors[c], label=sim)
    
    if c == 0:
        v_m_mrt2, v_chi_mrt2, v_te_mrt2, v_cv_mrt2 = v_m, v_chi, v_te, v_cv
        v_m_mrt2_err, v_chi_mrt2_err, v_te_mrt2_err, v_cv_mrt2_err = v_m_err, v_chi_err, v_te_err, v_cv_err
    else: 
        v_m_gibbs, v_chi_gibbs, v_te_gibbs, v_cv_gibbs = v_m, v_chi, v_te, v_cv
        v_m_gibbs_err, v_chi_gibbs_err, v_te_gibbs_err, v_cv_gibbs_err = v_m, v_chi, v_te, v_cv
for ax, title, ylabel in zip(
    [ax1, ax2, ax3, ax4],
    ["Magnetization", "Susceptibility", "Total Energy", "Specific Heat"],
    ["Magnetization", "Susceptibility", "Total Energy", "Specific Heat"]
):
    ax.set_title(title, fontsize=20)
    ax.set_xlabel("Temperature", fontsize=14)
    ax.set_ylabel(ylabel, fontsize=14)
    ax.legend(loc='best', fontsize=12)
    ax.grid(True)
```

```python
def chi_square(data, true_value, sigma):
    chi = 0
    for i in range(len(data)):
        chi += (pow(data[i] - true_value[i], 2) / sigma[i]**2)
    return chi / len(data)

T = np.linspace(0.5, 2, 7)
beta = 1 / T
th = np.tanh(J / T)
thN = th**Ns
ch = 1 / th
b = 1 / T
l1 = np.exp(b * J) * np.cosh(b * h) + np.sqrt(np.exp(2 * b * J) * np.cosh(b * h)**2 - 2 * np.sinh(2 * b * J))
l2 = np.exp(b * J) * np.cosh(b * h) - np.sqrt(np.exp(2 * b * J) * np.cosh(b * h)**2 - 2 * np.sinh(2 * b * J))
Z = l1**Ns + l2**Ns

v_m_true = (np.exp(b * J) * np.sinh(b * h) * 
    ((l1**(Ns - 1)) * (1 + np.exp(b * J) * np.cosh(b * h) / np.sqrt(np.exp(2 * b * J) * np.cosh(b * h)**2 - 2 * np.sinh(2 * b * J))) + 
     (l2**(Ns - 1)) * (1 - np.exp(b * J) * np.cosh(b * h) / np.sqrt(np.exp(2 * b * J) * np.cosh(b * h)**2 - 2 * np.sinh(2 * b * J)))) / Z)
v_chi_true = beta * np.exp(2 * beta * J) * (1 - thN) / (1 + thN)
v_cv_true = ((beta * J)**2) * (((1 + thN + (Ns - 1) * (th**2) + (Ns - 1) * (ch**2) * thN) / (1 + thN)) - Ns * ((th + ch * thN) / (1 + thN))**2)
v_te_true = -J * (th + ch * thN) / (1 + thN)

#print(v_chi_mrt2_err)

print("Chi^2 magnetization with Metropolis: ", chi_square(v_m_mrt2, v_m_true, v_m_mrt2_err))
print("Chi^2 susceptibility with Metropolis: ", chi_square(v_chi_mrt2, v_chi_true, v_chi_mrt2_err))
print("Chi^2 total energy with Metropolis: ", chi_square(v_te_mrt2, v_te_true, v_te_mrt2_err))
print("Chi^2 heat capacity with Metropolis: ", chi_square(v_cv_mrt2, v_cv_true, v_cv_mrt2_err))

print("\nChi^2 magnetization with Gibbs: ", chi_square(v_m_gibbs, v_m_true, v_m_gibbs_err))
print("Chi^2 susceptibility with Gibbs: ", chi_square(v_chi_gibbs, v_chi_true, v_chi_gibbs_err))
print("Chi^2 total energy with Gibbs: ", chi_square(v_te_gibbs, v_te_true, v_te_gibbs_err))
print("Chi^2 heat capacity with Gibbs: ", chi_square(v_cv_gibbs, v_cv_true, v_cv_gibbs_err))
```
